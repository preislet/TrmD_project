```python
import numpy as np
import pandas as pd
```

```python
# load docking_results.csv
df = pd.read_csv('docking_results.csv')

# sort by the score column
df = df.sort_values(by='Score')
```

```python
# print the top 10 rows
df.head(20)
```

```python
# We have rows with the same ligand, but different conformations
# We want the average score for each ligand
# We can use the groupby method to group by the ligand column
# CHEMBL3833409_preprocessed_confX_out.sdf

# split the ligand column by the underscore
df['Ligand'] = df['Ligand'].str.split('_').str[0]
# group the same ligands together and calculate the mean score
df = df.groupby('Ligand').mean()
# sort by the score column
df = df.sort_values(by='Score')
```

```python
# print the top 10 rows
print(df.head(50))
```

```python
# construct a new dataframe with the top 100 ligands
df_top = df.head(100)

# save the top 100 ligands to a new csv file
df_top.to_csv('top_100_ligands.csv')
```
